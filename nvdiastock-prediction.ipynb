```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

# ------------------------------
# Data Loading and Exploration
# ------------------------------
# Load the data with specified columns
data = pd.read_csv('/kaggle/input/nvidia-stock-data/NVIDIA_STOCK.csv', header=2, usecols=[0, 1, 2, 3, 4, 5, 6])

# Rename columns
data.columns = ['Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']

# Ensure 'Date' is a datetime object and set it as the index
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Print the first few rows to verify
print("Data loaded with new column names:")
print(data.head())

# Dataset Info and Description
print("\nDataset Info:")
print(data.info())
print("\nDataset Description:")
print(data.describe())

# Visualize Closing Price
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Adj Close'], label='Closing Price', color='blue')
plt.title('Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Decompose Time Series
decomposition = seasonal_decompose(data['Adj Close'], model='multiplicative', period=30)
decomposition.plot()
plt.show()

# Stationarity Test
result = adfuller(data['Adj Close'])
print("\nADF Test Results:")
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")
if result[1] < 0.05:
    print("The time series is stationary.")
else:
    print("The time series is not stationary.")

# Focus on 'Adj Close' column for prediction
prices = data['Adj Close'].values
prices = prices.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_prices = scaler.fit_transform(prices)

# Create sequences for LSTM/other models
sequence_length = 60  # Use 60 days of data to predict the next day
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_prices, sequence_length)

# Split data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Convert data to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)

# Create DataLoader for batching
train_data = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)

# Define the LSTM model in PyTorch (LSTM, GRU, CNN, Feedforward, BiLSTM)

# 1. LSTM Model
class LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_layer_size=50, output_size=1):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_layer_size, batch_first=True)
        self.dropout = nn.Dropout(0.2)
        self.fc1 = nn.Linear(hidden_layer_size, 25)
        self.fc2 = nn.Linear(25, output_size)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        lstm_out = self.dropout(lstm_out)
        lstm_out = lstm_out[:, -1, :]  # Get the output of the last time step
        x = self.fc1(lstm_out)
        x = self.fc2(x)
        return x

# 2. GRU Model
class GRUModel(nn.Module):
    def __init__(self, input_size=1, hidden_layer_size=50, output_size=1):
        super(GRUModel, self).__init__()
        self.gru = nn.GRU(input_size, hidden_layer_size, batch_first=True)
        self.dropout = nn.Dropout(0.2)
        self.fc1 = nn.Linear(hidden_layer_size, 25)
        self.fc2 = nn.Linear(25, output_size)

    def forward(self, x):
        gru_out, _ = self.gru(x)
        gru_out = self.dropout(gru_out)
        gru_out = gru_out[:, -1, :]  # Get the output of the last time step
        x = self.fc1(gru_out)
        x = self.fc2(x)
        return x

# Feedforward Neural Network
class FeedforwardNN(nn.Module):
    def __init__(self, input_size=60, hidden_layer_size=128, output_size=1):
        super(FeedforwardNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_layer_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_layer_size, output_size)

    def forward(self, x):
        x = x.view(x.size(0), -1)  # Flatten input data (batch_size, sequence_length)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# CNN Model
class CNNModel(nn.Module):
    def __init__(self, input_size=1, output_size=1):
        super(CNNModel, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=input_size, out_channels=64, kernel_size=3, padding=1)
        self.pool = nn.MaxPool1d(2)
        self.fc1 = nn.Linear(64 * (60 // 2), 50)  # Adjust the input size for the fully connected layer
        self.fc2 = nn.Linear(50, output_size)

    def forward(self, x):
        x = x.permute(0, 2, 1)  # Convert to (batch_size, channels, time_steps)
        x = self.conv1(x)
        x = self.pool(x)
        x = x.reshape(x.size(0), -1)  # Flatten for fully connected layers
        x = self.fc1(x)
        x = self.fc2(x)
        return x

# CNN + LSTM Model
class CNN_LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_layer_size=50, output_size=1):
        super(CNN_LSTMModel, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=input_size, out_channels=64, kernel_size=3, padding=1)
        self.pool = nn.MaxPool1d(2)
        self.lstm = nn.LSTM(64, hidden_layer_size, batch_first=True)
        self.fc1 = nn.Linear(hidden_layer_size, 25)
        self.fc2 = nn.Linear(25, output_size)

    def forward(self, x):
        x = x.permute(0, 2, 1)  # Convert to (batch_size, channels, time_steps)
        x = self.conv1(x)
        x = self.pool(x)
        x = x.permute(0, 2, 1)  # Convert back to (batch_size, time_steps, features) for LSTM
        lstm_out, _ = self.lstm(x)
        lstm_out = lstm_out[:, -1, :]  # Get the output of the last time step
        x = self.fc1(lstm_out)
        x = self.fc2(x)
        return x

# BiLSTM Model
class BiLSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_layer_size=50, output_size=1):
        super(BiLSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_layer_size, batch_first=True, bidirectional=True)
        self.dropout = nn.Dropout(0.2)
        self.fc1 = nn.Linear(hidden_layer_size * 2, 25)  # Multiply by 2 for bidirectional
        self.fc2 = nn.Linear(25, output_size)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        lstm_out = self.dropout(lstm_out)
        lstm_out = lstm_out[:, -1, :]  # Get the output of the last time step
        x = self.fc1(lstm_out)
        x = self.fc2(x)
        return x

# List of models to test
models = [
    LSTMModel(),
    GRUModel(),
    FeedforwardNN(input_size=sequence_length),
    CNNModel(),
    CNN_LSTMModel(),
    BiLSTMModel()
]

# Training and evaluation for each model
model_performance = {}

for model in models:
    print(f"\nTraining and evaluating {model.__class__.__name__}...")
    
    # Initialize loss function and optimizer
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    # Train the model
    epochs = 20
    train_losses = []

    for epoch in range(epochs):
        model.train()
        running_loss = 0
        for batch_X, batch_y in train_loader:
            optimizer.zero_grad()
            output = model(batch_X)
            loss = criterion(output, batch_y)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        avg_loss = running_loss / len(train_loader)
        train_losses.append(avg_loss)
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {avg_loss:.4f}")

    # Plot training loss
    plt.plot(train_losses, label=f'{model.__class__.__name__} - Training Loss')
    plt.legend()

    # Make predictions on test data
    model.eval()
    with torch.no_grad():
        predictions = model(X_test_tensor)
        predictions = predictions.numpy()

    # Inverse transform predictions and actual values
    predictions = scaler.inverse_transform(predictions)
    y_test_actual = scaler.inverse_transform(y_test_tensor.reshape(-1, 1))

    # Plot the predictions vs actual values
    plt.figure(figsize=(14, 5))
    plt.plot(y_test_actual, label='Actual Price', color='blue')
    plt.plot(predictions, label='Predicted Price', color='red')
    plt.title(f'{model.__class__.__name__} Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Calculate MSE for model performance
    mse = mean_squared_error(y_test_actual, predictions)
    model_performance[model.__class__.__name__] = mse

# Compare models based on MSE
print("\nModel Performance (MSE):")
for model_name, mse in model_performance.items():
    print(f"{model_name}: MSE = {mse:.4f}")

print("Analysis and Forecasting Complete")
```
